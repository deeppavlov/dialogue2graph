```python
from tqdm import tqdm
from augmentation_utils import check_no_duplicates_one_dialog
```

```python
from datasets import load_dataset

dataset = load_dataset("DeepPavlov/d2g_generated", token=True)
```

```python
errors = []
failure_examples = []
new_data = []

for i, example in enumerate(tqdm(dataset["train"])):
    print(f"example {i}")
    dialogs = [dial["messages"] for dial in example["dialogs"]]
    true_count = 0

    for j, dia in enumerate(dialogs):
        try:
            # if there're duplcates in the dialog, add graph serial number
            # and dialog serial number to the failure_examples
            if not check_no_duplicates_one_dialog(dia):
                failure_examples.append((i, j))
            else:
                true_count += 1
        except Exception as e:
            errors.append((i, j, e))

    # data without duplicates
    if true_count == len(dialogs):
        new_data.append(example)
```
